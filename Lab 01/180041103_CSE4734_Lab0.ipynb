```python
import numpy as np
import glob
import matplotlib.pyplot as plt
import skimage.io
import skimage.color
import skimage.filters


# load the image
image = skimage.io.imread("gray.jpg")

fig, ax = plt.subplots()
plt.imshow(image)   
plt.show()
```

```python
gray_image = skimage.color.rgb2gray(image)
# blur the image to denoise
blurred_image1 = skimage.filters.gaussian(gray_image, sigma=1.0)

fig, ax = plt.subplots()
plt.imshow(blurred_image1, cmap="gray")
plt.show()
```

```python
blurred_image1
```

```python
blurred_image1.shape
```

```python
histogram, bin_edges = np.histogram(blurred_image1, bins=256, range=(0.0, 1.0))
fig, ax = plt.subplots()
plt.plot(bin_edges[0:-1], histogram)
plt.title("Grayscale Histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixels")
plt.xlim(0, 1.0)
plt.show() 
```

# Explationation:
We know that histogram of an image represent the relative frequency of occurance of various gray label of an image. From the histogram we can see that the image is not so bright nor so dark, the range is in between o.22 to 0.8. this means that this image was not a bright image.


```python
t = [0.4, 0.5, 0.6, 0.7, 0.8, 0.9]
for i in t:
    binary_mask = blurred_image1 < i
    fig, ax = plt.subplots()
    plt.imshow(binary_mask, cmap="gray")
    plt.show()
```

# Explaination:
For different threshold value, we have different types of picture. If the threshold value is less then the picture becomes more dark because, we know, in case of thresholding if the intensity of a pixel is greater than threshold value then the pixel becomes white and rest of them remains black. When we increase the threshold value level, the image becomes more brighter than previous ones.

```python
# load the image
image = skimage.io.imread("bright.jpg")

fig, ax = plt.subplots()
plt.imshow(image)
plt.show()

gray_image = skimage.color.rgb2gray(image)
# blur the image to denoise
blurred_image2 = skimage.filters.gaussian(gray_image, sigma=1.0)

fig, ax = plt.subplots()
plt.imshow(blurred_image2, cmap="gray")
plt.show()

t = [0.4, 0.5, 0.6, 0.7, 0.8, 0.9]
for i in t:
    binary_mask = blurred_image2 < i
    fig, ax = plt.subplots()
    plt.imshow(binary_mask, cmap="gray")
    plt.show()
```
